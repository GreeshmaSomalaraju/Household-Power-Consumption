```python
df['DailyAvg']=df['DailyAvg'].fillna(method='ffill')
df['HourlyUsagePower']=df['HourlyUsagePower'].fillna(method='ffill')
```

```python
df['DailyAvg'] = df['DailyAvg'].astype(float)
df['HourlyUsagePower'] = df['HourlyUsagePower'].astype(float)
```

```python
from sklearn.model_selection import train_test_split

# Giving input and output(target)
X1=df[[
    'Global_reactive_power', 'Voltage', 'Global_intensity',
    'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3',
    'Hour', 'HourlyUsagePower', 'DailyAvg', 'rolling_1h'
] ]      # except target variable

Y1=df['Global_active_power']

# Split Train-Test
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y1, test_size=0.2, random_state=42)

# Training and Testing of Target Variable size
print('Training Set size',X1_train.shape)
print('Testing Set size',X1_test.shape)
```

```python
# Importing Important Libraries

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
```

```python
# Linear Regression

# Initilizing the ML model

lr_model=LinearRegression()

# Fitting the ML Model
lr_model.fit(X1_train,Y1_train)
```

```python
# Random Forest Regression

rf_model=RandomForestRegressor(n_estimators=100,random_state=42)
rf_model.fit(X1_train,Y1_train)
```

```python
# Gradient Boosting

gb_model=GradientBoostingRegressor(random_state=42)
gb_model.fit(X1_train,Y1_train)
```

```python
# Neural Network(MLP - Multi-Layer Perceptron)

nn_model = MLPRegressor(hidden_layer_sizes=(64, 64), max_iter=300, random_state=42)
nn_model.fit(X1_train, Y1_train)
```

```python
# Hyperparameter Tuning

from sklearn.model_selection import GridSearchCV

#  Random Forest Regressor
rf_params = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}
rf_grid = GridSearchCV(RandomForestRegressor(random_state=42), rf_params, cv=3, scoring='r2', n_jobs=-1)
rf_grid.fit(X1_train, Y1_train)
print(" Best RF Params:", rf_grid.best_params_)

#  Gradient Boosting Regressor
gb_params = {
    'n_estimators': [50, 100],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 5]
}
gb_grid = GridSearchCV(GradientBoostingRegressor(random_state=42), gb_params, cv=3, scoring='r2', n_jobs=-1)
gb_grid.fit(X1_train, Y1_train)
print(" Best GB Params:", gb_grid.best_params_)

#  Neural Network (MLPRegressor)
nn_params = {
    'hidden_layer_sizes': [(64,), (64, 64)],
    'activation': ['relu', 'tanh'],
    'learning_rate_init': [0.001, 0.01]
}
nn_grid = GridSearchCV(MLPRegressor(max_iter=300, random_state=42), nn_params, cv=3, scoring='r2', n_jobs=-1)
nn_grid.fit(X1_train, Y1_train)
print(" Best NN Params:", nn_grid.best_params_)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Function to evaluate a model
def evaluate_model(name, model, X1_test, Y1_test):
    y_pred = model.predict(X1_test)
    rmse = np.sqrt(mean_squared_error(Y1_test, y_pred))
    mae = mean_absolute_error(Y1_test, y_pred)
    r2 = r2_score(Y1_test, y_pred)
    print(f" {name} Performance:")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE : {mae:.4f}")
    print(f"  R  : {r2:.4f}\n")

# Evaluate all models
evaluate_model("Linear Regression", lr_model, X1_test, Y1_test)
evaluate_model("Random Forest", rf_grid.best_estimator_, X1_test, Y1_test)
evaluate_model("Gradient Boosting", gb_grid.best_estimator_, X1_test, Y1_test)
evaluate_model("Neural Network", nn_grid.best_estimator_, X1_test, Y1_test)
```

```python
# Store model results in a list
model_scores = []

# Add results for each model
for name, model in [
    ("Linear Regression", lr_model),
    ("Random Forest", rf_grid.best_estimator_),
    ("Gradient Boosting", gb_grid.best_estimator_),
    ("Neural Network", nn_grid.best_estimator_)
]:
    y_pred = model.predict(X1_test)
    rmse = np.sqrt(mean_squared_error(Y1_test, y_pred))
    mae = mean_absolute_error(Y1_test, y_pred)
    r2 = r2_score(Y1_test, y_pred)
    model_scores.append({
        'Model': name,
        'RMSE': rmse,
        'MAE': mae,
        'R2 Score': r2
    })

# Convert to DataFrame for easy comparison
results_df = pd.DataFrame(model_scores).sort_values(by='R2 Score', ascending=False)
print(" Model Performance Comparison:\n")
print(results_df)
```
